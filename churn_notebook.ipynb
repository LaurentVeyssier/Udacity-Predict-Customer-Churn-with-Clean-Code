```python
cat_columns = [
    'Gender',
    'Education_Level',
    'Marital_Status',
    'Income_Category',
    'Card_Category'                
]

quant_columns = [
    'Customer_Age',
    'Dependent_count', 
    'Months_on_book',
    'Total_Relationship_Count', 
    'Months_Inactive_12_mon',
    'Contacts_Count_12_mon', 
    'Credit_Limit', 
    'Total_Revolving_Bal',
    'Avg_Open_To_Buy', 
    'Total_Amt_Chng_Q4_Q1', 
    'Total_Trans_Amt',
    'Total_Trans_Ct', 
    'Total_Ct_Chng_Q4_Q1', 
    'Avg_Utilization_Ratio'
]
```

```python
# Encode Churn dependent variable : 0 = Did not churned ; 1 = Churned
df['Churn'] = df['Attrition_Flag'].apply(lambda val: 0 if val == "Existing Customer" else 1)
# Drop redudant Attrition_Flag variable
df.drop('Attrition_Flag', axis=1, inplace=True)
# Drop useless variable
df.drop('CLIENTNUM', axis=1, inplace=True)
```

# Perform Exploratory Data Analysis - EDA

```python
perform_eda(df)
```

# Feature engineering and preprocessing

```python
df = encoder_helper(df, cat_columns)
```

```python
keep_cols = ['Customer_Age', 'Dependent_count', 'Months_on_book',
             'Total_Relationship_Count', 'Months_Inactive_12_mon',
             'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
             'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt',
             'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio',
             'Gender_Churn', 'Education_Level_Churn', 'Marital_Status_Churn', 
             'Income_Category_Churn', 'Card_Category_Churn', 'Churn']

# train test split 
X_train, X_test, y_train, y_test = perform_feature_engineering(df[keep_cols])
```

```python
X_train.head()
```

# Train models

```python
# This cell may take up to 15-20 minutes to run
train_models(X_train, X_test, y_train, y_test)
```

```python
# Load trained models
rfc_model = joblib.load('./models/rfc_model.pkl')
lr_model = joblib.load('./models/logistic_model.pkl')
```

# Analyze feature importances

```python
explainer = shap.TreeExplainer(rfc_model)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, plot_type="bar")
```

```python
# Summary plot for Class = 1 (churned)
explainer = shap.TreeExplainer(rfc_model)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values[1], X_test)
```

```python
results_path = "./images/results"
feature_importance_plot(rfc_model, X_train, results_path)
```

```python
class_names = ['Not Churned', 'Churned']
plt.figure(figsize=(15, 5))
ax = plt.gca()
disp = plot_confusion_matrix(lr_model, X_test, y_test, display_labels=class_names, cmap=plt.cm.Blues, xticks_rotation='horizontal', colorbar=False, ax=ax)
# Hide grid lines
ax.grid(False)
plt.show()
```
